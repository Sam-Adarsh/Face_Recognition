```python
from skimage.io import imread,imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
Mike= os.listdir("D:/output/mike tyson")
```

```python
Maluma=os.listdir("D:/output/maluma")
```

```python
Jaden=os.listdir("D:/output/jaden smith")
```

```python
limit=50
mike_images=[None]*limit
j=0
for i in Mike:
    if(j<limit):
        mike_images[j]=imread("D:/output/mike tyson/"+i)
        j+=1
    else:
        break
```

```python
limit=50
maluma_images=[None]*limit
j=0
for i in Maluma:
    if(j<limit):
        maluma_images[j]=imread("D:/output/maluma/"+i)
        j+=1
    else:
        break
```

```python
limit=50
jaden_images=[None]*limit
j=0
for i in Jaden:
    if(j<limit):
        jaden_images[j]=imread("D:/output/jaden smith/"+i)
        j+=1
    else:
        break
```

```python
jaden_gray = [None]*limit
j=0
for i in Jaden:
    if(j<limit):
        jaden_gray[j]=rgb2gray(jaden_images[j])
        j+=1
    else:
        break
```

```python
img=pd.concat([mike_data,maluma_data])
```

```python
imagerec=pd.concat([img,jaden_data])
```

```python
imagerec
```

```python
from sklearn.utils import shuffle
```

```python
imr=shuffle(imagerec).reset_index()
```

```python
imr
```

```python
imrec=imr.drop(['index'],axis=1)
```

```python
imrec
```

```python
x=imrec.values[:,:-1]
```

```python
y=imrec.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf=svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
